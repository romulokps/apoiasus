```
import pandas as pd
```

```
df = pd.read_csv('municipal.txt', sep = "|")
df['dt_MesAnoReferencia'] = df['dt_MesAnoReferencia']//10000
df = df.rename(columns={'dt_MesAnoReferencia':'dt_mes'})
```

```
df2 = df[df['de_cargo'].str.contains('MEDICO', na=False)]
df2 = df2[df2['de_cargo'].str.contains('PSF', na=False)]
df2 = df2[df2['de_tipocargo'] != 'Inativos / Pensionistas']
df2 = df2[df2['dt_ano'] > 2019]
df3 = df2
df2.tail(30)
```

```
df3 = df2
df3 = df3.pivot_table(index=['no_Servidor','de_ugestora'], columns= ['dt_ano','dt_mes'], values='vl_vantagens')
n = df3.count(axis=1)
df3['media']= df3.sum(axis=1)/n
df3['n'] = n
df3 = df3[df3['n']>5]
df3 = df3.sort_values(by = 'media', ascending=False)
df3.head(35)
```
